# Data by Tax year

```python
asset_allocation_cumulative = data.groupby(by = "taxyr").mean()
```

```python
asset_allocation_cumulative
```

```python
cols = ['pct_pe', 'pct_cash', 'pct_hf', 'pct_re', 'pct_ra', 'pct_trusts', 'pct_coops', 'pct_public_equity', 'pct_fix_inc']
```

```python
for i in range(1, 9):
    curr_year = asset_allocation_cumulative.iloc[i]
    prev_year = asset_allocation_cumulative.iloc[i-1]
    for col in cols:
        pct_change = (curr_year[col] - prev_year[col])/(prev_year[col])
        print(f'Percent change of {col} in {2008+i}: {pct_change}')
```

# Data by NPO

```python
years_data = []

for name, asset_allocation_by_npo in data.groupby(by = "ein"):
    years_data.append(pd.DataFrame(asset_allocation_by_npo).taxyr.nunique())
    
print(f'Average number of years of data per NPO: {statistics.mean(years_data)}')

number_of_years_data = []
years = []
for i in range(10):
    number_of_years_data.append(years_data.count(i))
    years.append(i)
plt.bar(years, number_of_years_data)
plt.xlabel('Years of Data')
plt.ylabel('Number of NPOs')
```
